## Convolution

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from scipy import signal
```

```python
d, sr = librosa.load("auds/set1.wav",8000)
d1, sr = librosa.load("auds/set2.wav",8000)
y = signal.convolve(d,d1)
```

```python
def signal_plot(s,sr,title=None,ax=None):
    t=np.linspace(0,len(s)/sr,len(s))
    ax.plot(t,s,lw=1)
    ax.grid(True)
    ax.set_xlabel('Time(s)')
    ax.set_ylabel('Amplitude')
    ax.set_title(title)
```

```python
fig, ax = plt.subplots(3,1,figsize=(7,6),tight_layout=True)
ax  = ax.flat
titles = [
    "signal 1",
    "signal 2",
    'Covolution of signals'
]
for title,(i,s) in zip(titles,enumerate([d,d1,y])):
    signal_plot(s,sr,title,ax[i])
plt.show()
```

## Correlation

### Auto Correlation

```python
fig, ax = plt.subplots(2,1,figsize=(7,4),tight_layout=True)
ax  = ax.flat

signal_plot(s,sr,"Original Signal",ax[0])

ax[1].acorr(y,maxlags=20)
ax[1].set_title('Auto Correlation')
ax[1].set_xlabel("Lags")
ax[1].set_ylabel("Coefficents")
ax[1].grid(True)
plt.show()
```

### Cross Correlation

```python
d, sr = librosa.load("auds/sample1.wav",8000)
d1, sr = librosa.load("auds/sample2.wav",8000)
y = signal.correlate(d,d1)
```

```python
fig, ax = plt.subplots(3,1,figsize=(7,6),tight_layout=True)
ax  = ax.flat

signal_plot(d,sr,"Signal 1",ax[0])
signal_plot(d1,sr,"Signal 2",ax[1])

ax[2].plot(np.arange(-100,100),y[:200])
ax[2].set_title('Cross Correlation of signals')
ax[2].set_xlabel("Lags")
ax[2].set_ylabel("Coefficents")

plt.show()
```
