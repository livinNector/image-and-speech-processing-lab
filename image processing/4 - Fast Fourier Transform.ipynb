```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def plot_row(imgs,titles):
    n = len(imgs)
    plt.figure(figsize=(20,25))
    for i in range(n):
        plt.subplot(1,n,i+1)
        plt.imshow(imgs[i],cmap="gray")
        plt.title(titles[i])
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
imgs = [None]*5
imgs[0] = cv2.imread("imgs/barbara_gray.bmp", 0)
imgs[1] = np.fft.fft2(imgs[0])
imgs[2] = np.fft.fftshift(imgs[1])
imgs[3] = np.fft.ifftshift(imgs[2])
imgs[4] = np.fft.ifft2(imgs[3])

imgs[1] = np.log(1+np.abs(imgs[1]))
imgs[2] = np.log(1+np.abs(imgs[2]))
imgs[3] = np.log(1+np.abs(imgs[3]))
imgs[4] = np.abs(imgs[4])

titles = [
    "Original Image","Spectrum",
    "Centered Spectrum","Decentralized",
    "Processed Image"
]
plot_row(imgs,titles)
```
