# Sharpening and Smoothening

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def plot_diff(imgs,titles):
    n = len(imgs)
    plt.figure(figsize=(10,10))
    for i in range(n):
        plt.subplot(1,n,i+1)
        plt.imshow(imgs[i],cmap="gray")
        plt.title(titles[i])
        plt.axis("off")
        plt.tight_layout()
    plt.show()
```

```python
def sharpen(img):
    filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    return cv2.filter2D(img,-1,filter)
```

## Mean Blur and sharpen filter

```python
img = plt.imread("imgs/barbara_gray.bmp",1)
blur = cv2.blur(img,(5,5))
sharp = sharpen(img)
plot_diff(
    [blur,img,sharp],
    ["Blurred Image","Original Image","Sharpened Image"]
)
```

## Gaussian Blur and Bliateral Blur

```python
gaussian_blur = cv2.GaussianBlur(img,(5,5),0)
bilateral_blur = cv2.bilateralFilter(img,9,75,75)
plot_diff(
    [img,gaussian_blur,bilateral_blur],
    ["Original Image","Gaussian Blur","Bilateral Blur"]
)
```
