# Edge Detection methods

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def plot_diff(imgs,titles):
    n = len(imgs)
    plt.figure(figsize=(10,10))
    for i in range(n):
        plt.subplot(1,n,i+1)
        plt.imshow(imgs[i],cmap="gray")
        plt.title(titles[i])
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

## Sobel Edge Detection

```python
def sobel(img):
    filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_vert = cv2.filter2D(img,-1,filter)
    sobel_hori = cv2.filter2D(img,-1,np.fliplr(filter.T))
    return sobel_vert,sobel_hori
```

```python
img = plt.imread("imgs/barbara_gray.bmp",1)
sobel_v,sobel_h = sobel(img)
plot_diff(
    [img,sobel_v,sobel_h],
    ["Original Image","Vertical Edge","Horizontal Edge"]
)
```

## Canny Edge Detection

```python
canny_edges = cv2.Canny(img,100,100)
plot_diff(
    [img,canny_edges],
    ["Original Image","Edge using Canny edge detection"]
)
```
